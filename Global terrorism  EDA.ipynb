```python
df_final= gtd.drop(columns=['approxdate','related', 'resolution','latitude',	'longitude',"individual",'dbsource','extended',
 'INT_LOG',
 'INT_IDEO',
 'INT_MISC',
 'INT_ANY',
 'related',
 'attacktype2_txt','attacktype3_txt',"summary",'targsubtype1_txt','corp1',
 'target2',
 'natlty2',
 'natlty2_txt',
 'targtype3',
 'targtype3_txt',
 'targsubtype3',
 'targsubtype3_txt',
 'corp3',
 'target3',
 'natlty3',
 'natlty3_txt','targsubtype2',
 'targsubtype2_txt',
 'corp2',
 'targsubtype1',
'target1',
'natlty1',
'natlty1_txt','weapsubtype1',
'weapsubtype1_txt','nperps',
 'nperpcap',
 'claimed', 'scite1',
 'scite2',
 'nkillus',
 'nkillter',
 'crit3',
 'doubtterr',
 'nwoundus',
 'nwoundte',
 'property',
  'specificity','vicinity','crit1','crit2','multiple','attacktype1','targtype1','guncertain1',
 'weaptype1',
 ])
```

```python
df_final.isnull().sum()
```

Columns with greater than 50% null values will be dropped. The rest of the columns imputation will be attempted.

```python
for col in df_final.columns:
    print(col, f'{df_final[col].isnull().sum() / df_final.shape[0] * 100} % NULL')
```

```python
perc = 40.0
min_count =  int(((100-perc)/100)*df_final.shape[0] + 1)
mod_df= df_final.dropna( axis=1,
                thresh=min_count)
```

Looking at the unique values in each column gives an idea of which columns to drop and which columns to keep. 

```python
list(mod_df.columns.values)
```

```python
mod_df.head()
```

Performing  EDA of mod_df

```python
mod_df['casualities']=mod_df['nkill']+mod_df['nwound']
```

```python
print('Regions with Highest Terrorist Attacks:',mod_df['country_txt'].value_counts().index[0])
print('Maximum people killed in an attack are:',mod_df['nkill'].max(),'that took place in',mod_df.loc[mod_df['nkill'].idxmax()].city)
print("Year with the most attacks:",mod_df['iyear'].value_counts().idxmax())
print("Month with the most attacks:",mod_df['imonth'].value_counts().idxmax())
print("Most Attack Types:",mod_df['attacktype1_txt'].value_counts().idxmax())
```

```python
mod_df[["nkill","nwound"]].describe()
```
