Columns with greater than 50% null values will be dropped. The rest of the columns imputation will be attempted.

```python
for col in df_final.columns:
    print(col, f'{df_final[col].isnull().sum() / df_final.shape[0] * 100} % NULL')
```

```python
perc = 40.0
min_count =  int(((100-perc)/100)*df_final.shape[0] + 1)
mod_df= df_final.dropna( axis=1,
                thresh=min_count)
```

Looking at the unique values in each column gives an idea of which columns to drop and which columns to keep. 

```python
list(mod_df.columns.values)
```

```python
mod_df.head()
```

```python
mod_df=(mod_df.fillna({'nkill': 0, 'nwound': 0,'ishostkid':0}))
```

```python
mod_df.isnull().sum()
```

```python
mod_df.dropna() 
```

Performing 

```python
print("Number of Unique Values in Features")
print("====================================")
for col in mod_df.select_dtypes(include=["object"]).columns:
    print(col, f": {mod_df[col].unique().shape[0]}")
    print("====================================")
```

```python
mod_df['casualities']=mod_df['nkill']+mod_df['nwound']
```

```python
# Import label encoder
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
mod_df['provstate']=label_encoder.fit_transform(mod_df['provstate'])
mod_df['city']= label_encoder.fit_transform(mod_df['city'])
mod_df['weaptype1_txt']= label_encoder.fit_transform(mod_df['weaptype1_txt'])
mod_df['targtype1_txt']= label_encoder.fit_transform(mod_df['targtype1_txt'])
mod_df['gname']= label_encoder.fit_transform(mod_df['gname'])
mod_df['region_txt']= label_encoder.fit_transform(mod_df['region_txt'])
mod_df['country_txt']= label_encoder.fit_transform(mod_df['country_txt'])
mod_df['attacktype1_txt']= label_encoder.fit_transform(mod_df['attacktype1_txt'])
```

```python
# Separating the dependent and independent variables
y =mod_df['attacktype1_txt']
X = mod_df.drop('attacktype1_txt', axis = 1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFE, SelectKBest, f_classif, chi2,mutual_info_classif, SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
from sklearn.multioutput import ClassifierChain

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, LabelEncoder
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
preds_train= rf.predict(X_train)
preds_test= rf.predict(X_test)
```

```python
print(rf.feature_importances_)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('\nThe accuracy of random forest Classifier is {}%'.format(accuracy_score(y_test, preds_test)*100))
```
